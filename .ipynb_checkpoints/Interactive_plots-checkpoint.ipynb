Deduce stellar properties from spectral type (in particular the mass and, if possible, radius, luminosity and temperature). FYI, the standard procedure to deduce the mass of a star from 3 different methods is (here, just focus on getting the spectral-type - mass):
1. Spectral-type - mass: it is obtained by computing the mean and standard deviation of the evolutionary masses of apparently-single stars with similar spectral types. The masses of these stars can be computed for instance from their measured physical properties (e.g., luminosity, effective temperature, surface gravity) as input in the [BONNSAI Bayesian tool](https://www.astro.uni-bonn.de/stars/bonnsai/) ([Shneider+14](https://ui.adsabs.harvard.edu/abs/2014A&A...570A..66S/abstract)), which estimates the evolutionary mass using single-star evolution tracks (see [Schneider+18](https://ui.adsabs.harvard.edu/abs/2018A&A...618A..73S/abstract) for an example). 
2. Evolutionary mass: it is obtained by inputting the values of stellar luminosity and effective temperature in the BONNSAI tool.
3. Spectroscopic mass: it is obtained from the surface gravity (deduced from the width of the absorption lines in the spectrum) and stellar radius derived from spectroscopy. The stellar parameters and abundances are derived from a spectroscopic analysis (e.g. with the CMFGEN model atmosphere code, see section “spectral analysis” in supplementary information [Shenar+22](https://ui.adsabs.harvard.edu/abs/2022NatAs...6.1085S/abstract) for more details).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial import distance as ssd
import ipywidgets as widgets
from IPython.display import display

def actualizar_graficos(selected_class):
    x, y= 6, 4
    df_class = df_all_values[df_all_values['Class'] == selected_class]

    df_num = df_class.select_dtypes(include=['float64', 'int64']).copy()

    df_num_clean = df_num.dropna(axis=1).copy()

    correlation_matrix_kendall = df_num_clean.corr(method='kendall')

    num_objetos = len(df_num_clean)

    plt.figure(figsize=(x, y))
    sns.heatmap(correlation_matrix_kendall, annot=True, cmap='RdYlBu', center=0, fmt='.2f',
                linewidths=0.5, cbar_kws={'shrink': .75, 'label': 'Kendall correlation coefficient'})
    plt.title(f'Kendall Correlation Matrix (for {num_objetos} objects with no NaN values)')
    plt.subplots_adjust(bottom=0.2, left=0.2)
    plt.show()

    matriz_positiva = correlation_matrix_kendall[correlation_matrix_kendall > 0].fillna(0)
    distancia_positiva = 1 - matriz_positiva  
    vinculos_positivos = linkage(ssd.squareform(distancia_positiva), method='ward')

    plt.figure(figsize=(x, y))
    dendrogram(vinculos_positivos, labels=matriz_positiva.columns, leaf_rotation=45, leaf_font_size=10)
    plt.title('Dendrogram for Positive Correlation')
    plt.show()

    matriz_negativa = correlation_matrix_kendall[correlation_matrix_kendall < 0].fillna(0)  
    np.fill_diagonal(matriz_negativa.values, 1)
    distancia_negativa = 1 - np.abs(matriz_negativa)
    vinculos_negativos = linkage(ssd.squareform(distancia_negativa), method='ward')

    plt.figure(figsize=(x, y))
    dendrogram(vinculos_negativos, labels=matriz_negativa.columns, leaf_rotation=45, leaf_font_size=10)
    plt.title('Dendrogram for Negative Correlation')
    plt.show()


class_selector = widgets.Dropdown(
    options=df_all_values['Class'].unique(),
    value='sg',
    description='Clase:',
    disabled=False
)

interactive_output = widgets.interactive(actualizar_graficos, selected_class=class_selector)
display(interactive_output)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kendalltau

def kendall_bootstrap_analysis(df, col1, col2, n=100, alpha=0.05):
    """
    Realiza análisis de correlación de Kendall con bootstrapping para dos columnas de un DataFrame.
    
    Parámetros:
        df (pd.DataFrame): El DataFrame que contiene las columnas a analizar.
        col1 (str): Nombre de la primera columna.
        col2 (str): Nombre de la segunda columna.
        n (int): Número de iteraciones para bootstrapping. Por defecto, 100.
        alpha (float): Nivel de significancia para calcular el intervalo de confianza. Por defecto, 0.05 (95% IC).
        
    Retorna:
        None. Imprime resultados y genera gráficos.
    """
    # Eliminar valores nulos de las columnas seleccionadas
    df_filtered = df.dropna(subset=[col1, col2])
    
    # Calcular el coeficiente de correlación de Kendall para las variables originales
    tau, p_value = kendalltau(df_filtered[col1], df_filtered[col2])
    print(f"Coeficiente de Kendall original ({col1} vs {col2}): {tau:.4f}, p-valor: {p_value:.4f}")
    
    # Configurar bootstrapping
    bootstrap_taus = []
    
    # Realizar bootstrapping con reemplazo
    for _ in range(n):
        bootstrap_sample = df_filtered.sample(frac=1, replace=True)
        tau_bootstrap, _ = kendalltau(bootstrap_sample[col1], bootstrap_sample[col2])
        bootstrap_taus.append(tau_bootstrap)
    
    # Calcular estadísticas del bootstrapping
    mean_tau = np.mean(bootstrap_taus)
    std_tau = np.std(bootstrap_taus)
    lower_bound = np.percentile(bootstrap_taus, 100 * alpha / 2)  # Límite inferior del IC
    upper_bound = np.percentile(bootstrap_taus, 100 * (1 - alpha / 2))  # Límite superior del IC
    
    # Resultados
    print("\nResultados del Bootstrapping:")
    print(f"Media del coeficiente de Kendall: {mean_tau:.4f}")
    print(f"Desviación estándar: {std_tau:.4f}")
    print(f"Intervalo de confianza al {100 * (1 - alpha):.1f}%: [{lower_bound:.4f}, {upper_bound:.4f}]")
    
    # Verificar significancia
    if lower_bound > 0 or upper_bound < 0:
        print(f"El coeficiente es significativamente diferente de 0 al nivel del {100 * alpha:.1f}%.")
    else:
        print(f"El coeficiente no es significativamente diferente de 0 al nivel del {100 * alpha:.1f}%.")
    
    # Visualizar histograma de los coeficientes de Kendall obtenidos
    plt.figure(figsize=(8, 6))
    plt.hist(bootstrap_taus, bins=20, color='blue', alpha=0.7, edgecolor='black')
    plt.axvline(mean_tau, color='red', linestyle='--', label=f'Media: {mean_tau:.4f}')
    plt.axvline(lower_bound, color='green', linestyle='--', label=f'Límite inferior IC: {lower_bound:.4f}')
    plt.axvline(upper_bound, color='green', linestyle='--', label=f'Límite superior IC: {upper_bound:.4f}')
    plt.title(f'Distribución del coeficiente de Kendall ({col1} vs {col2}) (Bootstrapping)', fontsize=14)
    plt.xlabel('Coeficiente de Kendall', fontsize=12)
    plt.ylabel('Frecuencia', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()

columna1 = 'Mo'
columna2 = "Period"
kendall_bootstrap_analysis(df_all_values, columna1, columna2, n=100, alpha=0.05)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import kendalltau
import seaborn as sns
import matplotlib.pyplot as plt

def kendall_correlation_matrix_with_heatmap(df, n=100, alpha=0.05):
    """
    Calcula la matriz de correlación de Kendall con bootstrapping e incertidumbres 
    y genera un mapa de calor con colores divergentes para los valores medios.
    
    Parámetros:
        df (pd.DataFrame): DataFrame que contiene las columnas numéricas a analizar.
        n (int): Número de iteraciones para bootstrapping. Por defecto, 100.
        alpha (float): Nivel de significancia para calcular el intervalo de confianza. Por defecto, 0.05 (95% IC).
    """
    # Seleccionar columnas numéricas
    numeric_df = df.select_dtypes(include=[np.number])

    print(f"Número de filas restantes después de eliminar nulos: {len(numeric_df)}")
    
    # Inicializar estructuras para la matriz de correlación y sus incertidumbres
    cols = numeric_df.columns
    mean_correlation_matrix = pd.DataFrame(index=cols, columns=cols, dtype=float)
    detailed_correlation_matrix = pd.DataFrame(index=cols, columns=cols, dtype=object)
    
    # Calcular correlaciones
    for i, col1 in enumerate(cols):
        for j, col2 in enumerate(cols):
            if i <= j:  # Calcular solo para la mitad superior y diagonal
                # Correlación original
                tau, _ = kendalltau(numeric_df[col1], numeric_df[col2])
                
                # Bootstrapping para incertidumbres
                bootstrap_taus = []
                for _ in range(n):
                    sample = numeric_df.sample(frac=1, replace=True)
                    tau_bootstrap, _ = kendalltau(sample[col1], sample[col2])
                    bootstrap_taus.append(tau_bootstrap)
                
                # Estadísticas
                mean_tau = np.mean(bootstrap_taus)
                lower_bound = np.percentile(bootstrap_taus, 100 * alpha / 2)
                upper_bound = np.percentile(bootstrap_taus, 100 * (1 - alpha / 2))
                
                # Guardar resultados
                mean_correlation_matrix.at[col1, col2] = mean_tau
                detailed_correlation_matrix.at[col1, col2] = f"{mean_tau:.2f}+{upper_bound-mean_tau:.2f}-{mean_tau-lower_bound:.2f}"
                
                if i != j:
                    mean_correlation_matrix.at[col2, col1] = mean_tau
                    detailed_correlation_matrix.at[col2, col1] = detailed_correlation_matrix.at[col1, col2]
    
    # Visualización del mapa de calor
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        mean_correlation_matrix.astype(float),
        annot=True,
        fmt=".2f",
        cmap="RdBu",
        center=0,
        cbar_kws={"label": "Coeficiente de Kendall"}
    )
    plt.title("Mapa de calor de correlación de Kendall", fontsize=16)
    plt.tight_layout()
    plt.show()
    
    return detailed_correlation_matrix

# Usar la función con df_union
correlation_matrix_detailed = kendall_correlation_matrix_with_heatmap(df_all_values, n=100, alpha=0.05)

# Mostrar matriz de correlación detallada
print("\nMatriz de correlación detallada con incertidumbres:")
print(correlation_matrix_detailed)
```
