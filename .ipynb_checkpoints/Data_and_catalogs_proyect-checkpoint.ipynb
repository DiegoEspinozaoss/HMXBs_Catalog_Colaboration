## Here i show the amount of null and non-null values for each ID column for both catalogs. 

For what we see here, the most complete ID's are Main_ID and Name (Fortin/Neumann respectively). Also, the most incomplete ones are ROSAT_ID and UHURU4_ID

```python
import matplotlib.pyplot as plt

def hist(df):
    null_counts = df.isnull().sum()
    non_null_counts = df.notnull().sum()

    df_nulls = pd.DataFrame({
        'Nulls': null_counts,
        'Non-Nulls': non_null_counts
    })

    ax = df_nulls.plot(kind='bar', figsize=(10, 6), width=0.8)

    ax.set_title('Null and Non-Null Counts per Column')
    ax.set_ylabel('Count')
    ax.set_xlabel('Columns')

    plt.xticks(rotation=60, ha='right')
    plt.tight_layout()

    plt.show()
    
hist(cat_neuman)
hist(v2023_09_Fortin)
```

```python
import matplotlib.pyplot as plt
import numpy as np

null_counts_fortin = v2023_09_Fortin[columns_fortin].isnull().sum()
non_null_counts_fortin = v2023_09_Fortin[columns_fortin].notnull().sum()

null_counts_neuman = cat_neuman[columns_neuman].isnull().sum()
non_null_counts_neuman = cat_neuman[columns_neuman].notnull().sum()

fig, ax = plt.subplots(figsize=(10, 6))

indices_fortin = np.arange(len(columns_fortin))
indices_neuman = np.arange(len(columns_neuman)) + len(columns_fortin)

bar_width = 0.35

ax.bar(indices_fortin - bar_width/2, null_counts_fortin, bar_width, label='Nulls (Fortin)', color='orange')
ax.bar(indices_fortin + bar_width/2, non_null_counts_fortin, bar_width, label='Non-Nulls (Fortin)', color='cyan')

ax.bar(indices_neuman - bar_width/2, null_counts_neuman, bar_width, label='Nulls (Neuman)', color='red', alpha=0.7)
ax.bar(indices_neuman + bar_width/2, non_null_counts_neuman, bar_width, label='Non-Nulls (Neuman)', color='blue', alpha=0.7)

ax.set_xlabel('Columns')
ax.set_ylabel('Count of Values')
ax.set_title('Count of Null and Non-Null Values per Column')
ax.set_xticks(np.concatenate([indices_fortin, indices_neuman]))
ax.set_xticklabels(columns_fortin + columns_neuman, rotation=90)

ax.legend()

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

common_count_matrix = np.zeros((len(columns_fortin), len(columns_neuman)))

for i, fortin_col in enumerate(columns_fortin):
    fortin_objects = v2023_09_Fortin[fortin_col].dropna().unique()
    for j, neuman_col in enumerate(columns_neuman):
        neuman_objects = cat_neuman[neuman_col].dropna().unique()
        common_objects = set(fortin_objects) & set(neuman_objects)
        common_count_matrix[i, j] = len(common_objects)

common_count_df = pd.DataFrame(common_count_matrix, index=columns_fortin, columns=columns_neuman)

plt.figure(figsize=(12, 8))
sns.heatmap(common_count_df, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Common Objects Count'}, fmt=".0f")

plt.title('Heatmap of Common Objects Between Columns of Fortin and Neuman', fontsize=16)
plt.xlabel('Neuman Columns', fontsize=14)
plt.ylabel('Fortin Columns', fontsize=14)

plt.tight_layout()
plt.show()
```

## Here i obtain all the systems which are either in Neumann or in Fortin Catalog, doing a cross-match of all the id's columns. 

```python
import pandas as pd

unique_fortin_objects = set()
for fortin_col in columns_fortin:
    fortin_objects = v2023_09_Fortin[fortin_col].dropna().unique()
    unique_fortin_objects.update(fortin_objects)

unique_neuman_objects = set()
for neuman_col in columns_neuman:
    neuman_objects = cat_neuman[neuman_col].dropna().unique()
    unique_neuman_objects.update(neuman_objects)

common_unique_objects = unique_fortin_objects & unique_neuman_objects

fortin_main_ids = v2023_09_Fortin['Main_ID'].dropna().unique()

common_fortin_objects = [obj for obj in fortin_main_ids if obj in common_unique_objects]

common_objects_data = []

for obj in common_fortin_objects:
    matching_columns = []
    for fortin_col in columns_fortin:
        if v2023_09_Fortin[fortin_col].dropna().isin([obj]).any():
            for neuman_col in columns_neuman:
                if cat_neuman[neuman_col].dropna().isin([obj]).any():
                    matching_columns.append(f"{fortin_col} - {neuman_col}")
    if matching_columns:
        common_objects_data.append({"object": obj, "coincidencia": ", ".join(matching_columns)})

common_id_objects = pd.DataFrame(common_objects_data)

common_id_objects
id_objects = common_id_objects['object']
id_objects
```

## Here i merge the id_objects and coord_objects to be more rigorous:

```python
all_unique_objects = list(set(id_objects) | set(coord_objects))

print("All unique objects from both lists:")
len(all_unique_objects)
```

```python
from matplotlib_venn import venn2
from matplotlib import pyplot as plt

id_objects_set = set(id_objects)
coord_objects_set = set(coord_objects) 

venn2(
    subsets=(len(id_objects_set - coord_objects_set), 
             len(coord_objects_set - id_objects_set), 
             len(id_objects_set & coord_objects_set)),
    set_labels=("ID Objects", "Coord Objects")
)

plt.title("Venn Diagram for Id and Coord objects")
plt.show()
```

## Here i combine the Fortin and Neumann Catalog.

A little problem here: i can't get all the systems whose names are defined either by the id or by the equatorial coordinates

```python
cols_f = ['Main_ID', 'Spectype', 'Mo', 'Period', 'Eccentricity', 'Spin_period', 'Distance', 'Class']
cols_n = ['Name', 'SpType', 'Mean_Mass', 'Teff', 'N_H', 'Max_Soft_Flux', 'Min_Soft_Flux', 'Max_Hard_Flux', 'Min_Hard_Flux']

fortin_filtered = v2023_09_Fortin[v2023_09_Fortin['Main_ID'].isin(all_unique_objects)][cols_f]
neuman_filtered = cat_neuman[cat_neuman['Name'].isin(all_unique_objects)][cols_n]

fortin_filtered = fortin_filtered.rename(columns={'Main_ID': 'Object'})
neuman_filtered = neuman_filtered.rename(columns={'Name': 'Object'})

df_final = pd.merge(fortin_filtered, neuman_filtered, on='Object', how='outer')

print("DataFrame resultante con columnas adicionales:")
df_final
```

```python
df_final['Mean_Soft_Flux'] = np.sqrt(df_final['Min_Soft_Flux'] * df_final['Max_Soft_Flux'])

df_final['Mean_Hard_Flux'] = np.sqrt(df_final['Min_Hard_Flux'] * df_final['Max_Hard_Flux'])

df_final['Hardness'] = df_final['Mean_Hard_Flux'] / df_final['Mean_Soft_Flux']
```

```python
df_final.rename(columns={"SpType": "SpType_Fortin", "Spectype": "SpType_Kim", "Mean_Mass": "M_X", "Mo":"M*"},
                inplace=True)
```
