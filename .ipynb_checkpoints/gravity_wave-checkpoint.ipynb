### Model
The sinusoidal waves at the surface $z=h$ of the pool correspond to a plane progressive wave characterized by:
- a wavelength $\lambda$ and a wave number $k=2\pi/\lambda$
- a period $T$ and a pulsation $\omega=2\pi/T$
- a velocity $c=\lambda/T=\omega/k$<br>
We found the velocity components (in Cartesian coordinates):
$$
v_x(x,z)=kA\cosh (kz) \sin (\omega t -kx)
$$
$$
v_z(x,z)=kA\sinh (kz) \cos (\omega t -kx)
$$
with $A$ an unknown constant (necessarily homogeneous to a length squared divided by a time). We use $1/k$ and $1/\omega$ as the length and time unit respectively in order to write the velocity in their dimensionless form:
$$
v_x(x,z)=A\cosh (z) \sin (t -x)
$$
$$
v_z(x,z)=A\sin (z) \cos (t -x)
$$
where the velocities are in unit of $c$ and $A$ is in unit of $\omega/k^2$. In this unit system, we have:
$$
\lambda=2\pi \quad \text{\&} \quad T=2\pi
$$
and from now on, we work exclusively in dimensionless units.

```python
lmbd=2*np.pi
T=2*np.pi
```

The only two degrees-of-freedom which are going to define the shape of the solutions (i.e. the shape parameters, in contrast with the scale parameters which only play a role when we come back to physical units) are $A$ and the depth $h$ of the pool. We can replace the shape parameter $A$ with $a(z=h)=a_h=A\cosh(h)$, the semi-major axis of the fluid particles' trajectories at the top of the pool. Changing any other parameter of the problem while keeping $A$ and $h$ identic will not modify the shape of the solutions. To ensure that the hypothesis of the problem are respected, we need $h$ and $a$ to be respectively very large and very small compared to the wavelength:
$$
a_h<<2\pi<<h
$$
In these conditions, we have seen that the trajectory of a fluid particle of time-averaged position $(x_m,z_m)$ is given by:
$$
x(t)=x_m-A\cosh (z_m) \cos (t -x_m)
$$
$$
z(t)=z_m+A\sinh (z_m) \sin (t -x_m)
$$
It is an ellipse of parameters $a=A\cosh (z_m)$ and $b=A\sinh (z_m)<a$ whose aspect ratio drops once $z_m$ becomes smaller than $1$ (due to the shape of the function $\tanh$). Given how fast the size of the trajectories shrinks as we go deeper in the pound (since $\cosh(x)\sim e^x$ for $x>>1$), we choose to verify the inequality above only marginally hereafter, and we set:
- $a_h=0.3\times 2\pi$ (since the semi-major axis of the fluid particles' trajectories decreases as $z$ decreases, $a<<2\pi$ is verified $\forall z$)
- $h=3\times 2\pi$ 

### Velocity components
Here, argument $A$ doesn't play any role because we don't need it to plot the streamlines

```python
def get_v(x,z,t,A): 
    vx = A * np.cosh(z) * np.sin(t-x)
    vz = A * np.sinh(z) * np.cos(t-x)
    return vx,vz
```

Alternatively, we can plot directly the streamlines using the analytic equation we found:
$$
\sinh (z) \sin (t-x) = C(t)
$$
with $C(t)$ a value which depends on time but which can be found by setting the seeds of the streamlines, that is to say by looking for the streamlines which go through pre-set points $(x_0,z_0)$ and computing the $C(t)$ associated to each of these points.

```python
def strml_cst(t,x0,z0): # for a given seed (x0,z0) @ a specific time t, find the corresponding constant
    C=np.sinh(z0)*np.sin(t-x0)
    return C
```

```python
def strml_eq(t,x,C): 
    z=np.arcsinh(C/np.sin(t-x))
    return z
```

### Position of fluid particles

```python
def get_pos(t,xm,zm,A):
    x = xm - A * np.cosh(zm) * np.cos(t-xm)
    z = zm + A * np.sinh(zm) * np.sin(t-xm)
    return x,z
```

### Equation of the upper surface

```python
def surf(t,x,h,A):
    z = h + A * np.sinh(h)*np.sin(t-x)
    return z
```

### Physical parameters

In order to limit the gap between scales, we choose $h=10$ times the semi-major axis of the trajectories near $z=h$.

```python
# SHAPE PARAMETERS ================================================
ah=0.3*lmbd # maximum semi-major axis of trajectories (i.e. @ z=h)
h=10*ah # such as h >> 2pi
# =================================================================
A=ah/np.cosh(h) # such as a=ah @ z=h
zps=[0.025*lmbd,0.88*h] # positions where we want to plot fluid particles' trajectories
print(A*np.cosh(zps[0])) # semi-major axis of the trajectories at the bottom --> TINY!
```

### Numeric parameters
Here, we set the boundaries such as we can capture 4 wavelengths $\lambda$, centered on $x=0$ => x_max-x_min$=4\lambda=4\times 2\pi$ in units of $1/k$.

```python
x_mx= 2*lmbd
x_mn=-2*lmbd
z_mx= 1.1*h
z_mn= 0

# for inset (i.e. at the bottom)
x2_mn=0-2*A*np.cosh(zps[0])
x2_mx=0+2*A*np.cosh(zps[0])
z2_mn=zps[0]-(x2_mx-x2_mn)/2.
z2_mx=zps[0]+(x2_mx-x2_mn)/2.

# grid resolution
Nx=30
Nz=30
```

```python
Z,X = np.mgrid[z_mn:z_mx:(Nz*1j),x_mn:x_mn+1.5*lmbd:(Nx*1j)] # we make the (reduced) Cartesian mesh (i.e. the grid) for streamlines
x_surf = np.linspace(x_mn,x_mx,100) 
```

**Caption of the plot below -** The slider enables to change time across one period. In blue the free surface between the fluid and the air, in black the trajectory of the fluid particles (independent of time) and in red the position of the fluid particle (both dependent on time). In orange, we represented (i) the streamlines deduced directly by Python's streamplot command from the components of the velocity field (solid lines) and (ii) the streamlines obtained from the analytic expression (semi-transparent dashed lines, less computationally demanding). We see that at a given time, the motion of the free surface is coherent with the streamlines: the component of the velocity normal to the free surface does not cancel out because it is not a fixed obstacle (but rather a free boundary). The inset illustrates the flatening of the trajectories once $z$ becomes smaller than $\lambda$, which is associated to the vertical stretching of the streamlines near the bottom. The dashed line at the bottom locates the $z$ of the inset.

```python
fig = plt.figure(figsize=(10,7*(x_mx-x_mn)/(z_mx-z_mn))) # to make it equal aspect ratio initially but allow to zoom
ax = fig.add_subplot(1,1,1)
# ax.set_aspect('equal')

def update(t=(0,T,0.1*T)):

    ax.clear()  

    ax.plot(x_surf,surf(t,x_surf,h,A)) # plot the upper free surface

    # Plot the streamlines ========================
    ## 1st method ---------------------------------
    # 40 (in z) times 2 (in x) seeds 
    for z in np.linspace(0,h+A*np.sinh(h),40): 
        for x in [x_mn+lmbd/4,x_mn+3*lmbd/4]: 
            C=strml_cst(t,x,z)
            zs=strml_eq(t,x_surf,C)
            zs = np.where(zs>surf(t,x_surf,h,A),np.nan,zs) # mask values above surface
            ax.plot(x_surf,zs,c='orange',ls='--',alpha=0.3)
    ## 2nd method ---------------------------------
    vx, vz = get_v(X,Z,t,1) # dumb A to plot streamlines so we set it to 1
    vx = np.where(Z>surf(t,X,h,A),np.nan,vx) # mask values above surface
    ax.streamplot(X,Z,vx,vz,broken_streamlines=False,density=[0.3,0.4])

    # Fluid particle @ higher z (=> bigger and more circular)
    zp=zps[1]
    a=A*np.cosh(zp)
    b=A*np.sinh(zp)
    ellipse = Ellipse((0,zp),2*a,2*b,alpha=1,ec='k',fc='None')
    ax.add_artist(ellipse)
    xp,yp = get_pos(t,0,zp,A)
    ax.scatter(xp,yp,s=140,marker='+',c='r')

    # Fluid particle @ lower z (=> smaller and flatter)
    ax.hlines(zps[0],x_mn,x_mx,ls='--',color='k')
    # inset =============================
    left, bottom, width, height = [0.55, 0.29, 0.4, 0.4] # These are in unitless percentages of the figure size. (0,0 is bottom left)
    ax2 = fig.add_axes([left, bottom, width, height])
    ax2.set_aspect('equal')
    zp=zps[0]
    a=A*np.cosh(zp)
    b=A*np.sinh(zp)
    ellipse = Ellipse((0,zp),2*a,2*b,alpha=1,ec='k',fc='None')
    ax2.add_artist(ellipse)
    xp,yp = get_pos(t,0,zp,A)
    ax.scatter(xp,yp,s=140,marker='+',c='r')
    ax2.scatter(xp,yp,s=140,marker='+',c='r')
    
    ax2.set_xlim(x2_mn,x2_mx)
    ax2.set_ylim(z2_mn,z2_mx) 
    ax2.set_title(r'$z={zp}\lambda\sim{zpp}$'.format(zp=zp/lmbd,zpp=np.round(zp,3)))
    
    ax.set_xlim(x_mn,x_mx)
    ax.set_ylim(z_mn,z_mx)    
   
    fig.canvas.draw_idle()

interact(update);
```
